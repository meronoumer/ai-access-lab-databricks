
# AI Access Lab Workshop Plan

## Workshop Title

From Messy Data to Responsible AI: A Beginner Databricks Workshop

## Audience

Beginner and intermediate Smith students who are curious about AI, machine learning, data science, or responsible technology, including students who may not yet have access to formal ML coursework.

## Learning Goals

By the end of the workshop, students will be able to:

1. Explain what happens before a machine learning model is trained.
2. Load and inspect a dataset in Databricks.
3. Use SQL to answer basic questions from data.
4. Understand why data cleaning and validation matter.
5. Train a simple model and interpret evaluation results.
6. Discuss limitations, bias, privacy, and responsible use.

## Why This Matters

Many students encounter AI through hype, fear, or misinformation. This workshop makes AI more concrete and less intimidating by showing the full workflow: data, cleaning, analysis, modeling, evaluation, and reflection.

## 60-Minute Workshop Structure

### 0–10 minutes: What is Databricks and why does data infrastructure matter?

Introduce Databricks as a platform for working with data, analytics, and AI workflows.

### 10–25 minutes: Explore the dataset

Students inspect the AI Access Lab dataset and discuss what each column means.

### 25–40 minutes: Find insights with SQL

Students answer questions like:
- What barriers do students report most often?
- Which topics are students most interested in?
- How does confidence vary by prior coding experience?

### 40–50 minutes: Train a simple model

Students see a simple model that predicts preferred workshop format.

### 50–60 minutes: Responsible AI discussion

Students discuss why the model should be interpreted carefully and how privacy, consent, and student agency matter.

## Campus Implementation Plan

I would use this project as the first technical workshop for the AI/ML club I am building at Smith. The goal would be to create a welcoming entry point for students who are curious about AI but may feel intimidated by coding, lack access to ML courses, or have mostly encountered AI through misinformation or fear.

As an AEMES Peer Mentor, calculus TA, and student coordinator, I already care deeply about peer teaching and campus community-building. This project gives me a concrete way to connect that work with Databricks: helping students move from curiosity to hands-on confidence.
