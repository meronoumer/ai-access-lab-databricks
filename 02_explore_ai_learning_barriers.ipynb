
# Exploring AI/ML Learning Barriers

This notebook uses PySpark and SQL to analyze student confidence, learning barriers, topic interests, and preferred workshop formats.

The goal is to turn messy student-learning data into insights that could guide campus programming.

```python
df = spark.table("ai_access_lab_students")
display(df)
```

```python
df.createOrReplaceTempView("students")
```

```python
%sql
SELECT confidence_group, COUNT(*) AS student_count
FROM students
GROUP BY confidence_group
ORDER BY student_count DESC;
```

```python
%sql
SELECT main_learning_barrier, COUNT(*) AS student_count
FROM students
GROUP BY main_learning_barrier
ORDER BY student_count DESC;
```

```python
%sql
SELECT
  beginner_status,
  preferred_workshop_format,
  COUNT(*) AS student_count
FROM students
GROUP BY
  beginner_status,
  preferred_workshop_format
ORDER BY
  beginner_status,
  student_count DESC;
```

## Key Insights

From this analysis, an AI/ML club could make more informed programming decisions:

1. If many students report low confidence, the club should not start with advanced model-building.
2. If many students say they do not know where to start, the first event should be a welcoming orientation.
3. If students are worried about AI ethics or misinformation, responsible AI should be built into the curriculum rather than treated as an optional topic.
4. If beginner students prefer hands-on project nights, the club can design low-pressure mini-projects instead of lecture-heavy events.
